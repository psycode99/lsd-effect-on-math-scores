```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('9.1 lsd_math_score_data.csv.csv')
df
```

```python
time = df[['Time_Delay_in_Minutes']]
lsd = df[['LSD_ppm']]
score = df[['Avg_Math_Test_Score']]
```

```python
%matplotlib inline

plt.figure(figsize=(14, 7), dpi=100)
plt.title('Tissue Concentration of LSD Over Time', fontsize=20)
plt.plot(time, lsd, c='#e74c3c', linewidth=2.5)
plt.xlabel('Time in Minutes', fontsize=14)
plt.ylabel('LSD ppm', fontsize=14)
plt.xlim(0, 500)
plt.ylim(1, 7)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.text(x=0, y=-0.5, s='Wagner et al. (1963)', fontsize=14)
plt.style.use('ggplot')
plt.show()
```

```python
regr = LinearRegression()
regr.fit(lsd, score)
```

```python
intercept = regr.intercept_[0]
print(f'The Intercept: {(intercept)}')
```

```python
coef = regr.coef_[0][0]
print(f'The co-efficient: {coef}')
print(f"Meaning that for every increase in LSD usage, the candidate's score gets worse by {coef}")
```

```python
rss = regr.score(lsd, score)*100
print(f'R-Squared: {round(rss)}%')
predicted_score = regr.predict(lsd)
```

```python
%matplotlib inline

plt.figure(figsize=(14,7), dpi=100)
plt.title('Effect of LSD on Math Score', fontsize=20)
plt.xlabel('LSD')
plt.ylabel('Performance Score')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlim(1, 6.5)
plt.ylim(25, 80)
plt.scatter(lsd, score, alpha=0.6, s=100.1)
plt.plot(lsd, predicted_score)
plt.style.use('fivethirtyeight')
plt.show()
```
